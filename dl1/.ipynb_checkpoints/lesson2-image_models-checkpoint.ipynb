## Multi-label versus single-label classification

```python
from fastai.plots import *
```

```python
def get_1st(path): return glob(f'{path}/*.*')[0]
```

```python
dc_path = "data/DogVsCats/valid/"
list_paths = [get_1st(f"{dc_path}cats"), get_1st(f"{dc_path}dogs")]
plots_from_files(list_paths, titles=["cat", "dog"], maintitle="Single-label classification")
```

In single-label classification each sample belongs to one class. In the previous example, each image is either a *dog* or a *cat*.

```python
list_paths = [f"{PATH}train-jpg/train_0.jpg", f"{PATH}train-jpg/train_1.jpg"]
titles=["haze primary", "agriculture clear primary water"]
plots_from_files(list_paths, titles=titles, maintitle="Multi-label classification")
```

In multi-label classification each sample can belong to one or more classes. In the previous example, the first images belongs to two classes: *haze* and *primary*. The second image belongs to four classes: *agriculture*, *clear*, *primary* and  *water*.

在多标签分类问题下，每个样本可以属于一个或多个类别。
比如第一章图属于两个类别“haze”和“primary”，第二张图属于四个类别“agriculture”， “clear”， “primary”和“water”.

## Multi-label models for Planet dataset

```python
from planet import f2

# 衡量方法 f2
# 常用的metrics是accuracy
metrics=[f2]

# 模型取resnet34
f_model = resnet34
```

```python
# csv路径
label_csv = f'{PATH}train_v2.csv'

# 获取训练集样本数目
n = len(list(open(label_csv)))-1

# 分割训练集和验证集，生成验证集序号
val_idxs = get_cv_idxs(n)
```

We use a different set of data augmentations for this dataset - we also allow vertical flips, since we don't expect vertical orientation of satellite images to change our classifications.

```python
# fast.ai 自动处理多分类数据的多个标签的读取
def get_data(sz):
    # 此处sz是做dataloader时将图像预处理为指定sz大小
    tfms = tfms_from_model(f_model, sz, aug_tfms=transforms_top_down, max_zoom=1.05)
    return ImageClassifierData.from_csv(PATH, 'train-jpg', label_csv, tfms=tfms,
                    suffix='.jpg', val_idxs=val_idxs, test_name='test-jpg')
```

```python
data = get_data(256)
```

```python
# data.val_dl 属于dataloader,返回的是mini-batch.因此，y是64x17，即batchsize为64，labels长度为17
x,y = next(iter(data.val_dl))
```

```python
# 在fast.ai和pytorch中，都是用hot-code来进行编码的。

list(zip(data.classes, y[0]))
```

```python
sz=64
```

```python
data = get_data(sz)
```

Planet的数据和ImageNet的数据不同，因此需要重新训练。

```python
# 此处是将dataloader中的图像放大1.3倍，便捷的方式来生成其他大小的数据图像
data = data.resize(int(sz*1.3), 'tmp')
```

```python
learn = ConvLearner.pretrained(f_model, data, metrics=metrics)
```

```python
lrf=learn.lr_find()
learn.sched.plot()
```

```python
lr = 0.2
```

```python
learn.fit(lr, 3, cycle_len=1, cycle_mult=2)
```

```python
# differencial learning rate的分配，除了第三个参数是针对全连接层的，第一二个参数平均分配剩余测网络层
lrs = np.array([lr/9,lr/3,lr])
```

```python
# 从新训练
# 事实上，可以直接跳过从预训练模型中训练全连接层参数这一步，直接unfreeze进行全部权重的训练
# 但是，预训练模型的卷积层参数不是随机的，而是在ImageNet数据集上训练过的。
#（对于与ImageNet类似的数据，是好事，对于与ImageNet不同的数据集，也比没有要好）
# 但是全连接层又是全部随机的，所以对于全连接层，可以先进行一次训练（在预训练卷积参数的帮助下），使之比随机要好，
# 然后再放开卷积层的参数，对其进行全部的训练
learn.unfreeze()
learn.fit(lrs, 3, cycle_len=1, cycle_mult=2)
```

```python
learn.save(f'{sz}')
```

```python
learn.sched.plot_loss()
```

```python
sz=128
```

```python
learn.set_data(get_data(sz))
learn.freeze()
learn.fit(lr, 3, cycle_len=1, cycle_mult=2)
```

```python
learn.unfreeze()
learn.fit(lrs, 3, cycle_len=1, cycle_mult=2)
learn.save(f'{sz}')
```

```python
sz=256
```

```python
learn.set_data(get_data(sz))
learn.freeze()
learn.fit(lr, 3, cycle_len=1, cycle_mult=2)
```

```python
learn.unfreeze()
learn.fit(lrs, 3, cycle_len=1, cycle_mult=2)
learn.save(f'{sz}')
```

```python
multi_preds, y = learn.TTA()
preds = np.mean(multi_preds, 0)
```
